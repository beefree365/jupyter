```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Optional: suppress font-related warnings
warnings.filterwarnings("ignore")

# Load your dataset
df = pd.read_csv("sample.csv")  # Or use pd.read_clipboard() to paste inline data

# Set crossing threshold
threshold = 0.1  # Only count crossing if y change exceeds this threshold

# Function: find valid crossings and return coordinates
def find_crossings(group_df, threshold):
    group_df = group_df.sort_values("x")
    y_mean = group_df["y"].mean()
    y_vals = group_df["y"].values
    x_vals = group_df["x"].values

    crossings = []
    for i in range(len(y_vals) - 1):
        y1, y2 = y_vals[i], y_vals[i + 1]
        # Detect crossing over mean
        if (y1 - y_mean) * (y2 - y_mean) < 0:
            if abs(y2 - y1) > threshold:
                # Linearly interpolate crossing point
                x_cross = x_vals[i] + (x_vals[i+1] - x_vals[i]) * abs(y1 - y_mean) / abs(y2 - y1)
                crossings.append((x_cross, y_mean))
    return crossings

# Get all groups and plot
groups = df["group"].unique()
fig, axes = plt.subplots(len(groups), 1, figsize=(10, 3.5 * len(groups)), dpi=100)
if len(groups) == 1:
    axes = [axes]

for ax, group in zip(axes, groups):
    group_df = df[df["group"] == group].copy()
    crossings = find_crossings(group_df, threshold)
    y_mean = group_df["y"].mean()

    # Plot raw data
    ax.plot(group_df["x"], group_df["y"], "o-", alpha=0.5, label=f"group {group}")

    # Plot mean line
    ax.axhline(y_mean, color="gray", linestyle="--", linewidth=1, label=f"Mean = {y_mean:.2f}")

    # Plot crossing points
    if crossings:
        cross_x, cross_y = zip(*crossings)
        ax.plot(cross_x, cross_y, "ro", label=f"Crossings ({len(crossings)} times)", markersize=6)

    ax.set_title(f"Group {group}: Mean Crossing Count (Threshold = {threshold})")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.tight_layout(h_pad=2.0)
plt.show()
```
